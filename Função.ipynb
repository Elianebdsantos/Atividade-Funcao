(4) Implemente uma função que receba um valor em segundos e imprima o correspondente em horas, minutos e segundos.

```python
def conver_temp(segundos):
  horas = segundos // 3600 # divisão dos segundos para horas, a // indica retorno da divisão inteira
  minutos= segundos // 60 #transformação dos minutos restantes em divisão inteira
  segundos= segundos % 60 #atualizar o restante do número incluindo somente os segundos, ou seja, já foi retirada as horas

  return f"{horas}:{minutos:02d}:{segundos:02d}" #indica o retorno do dado com a formatação de dois digitos, caso necessite vai preencher 0 a esquerda
```

```python
print(conver_temp(3890))
```

(7) Implemente a função todos_iguais(sequencia) que retorne True se todos os elementos da sequencia forem iguais, e retorne False caso contrário.

```python
def todos_iguais(n):
  if not n:
    return False
  primeiro_elemento = n[0]
  for elemento in n:
      if elemento != primeiro_elemento:
        return False
  return True
```

```python
print(todos_iguais([1,1,2]))
```

(8) Implemente a função todos_diferentes(sequencia) que retorne True se todos os elementos da sequencia forem diferentes entre si, e retorn False caso contrário, isto é, pelo menos um elemento seja igual a outro componente da sequencia



```python
def todos_diferentes(n):
  for i in range(len(n)):
    for j in range(i + 1, len(n)):
      if n [i] == n[j]:
        return False
```

```python
todos_diferentes([1,2,3,4,3])
```

(9) Implemente uma função que receba como argumento um número n e uma lista de números, indice_elemento(n, lista). A função retornará o índice da primeira ocorrência do número da lista e -1 se o número não estiver na lista. Obs. Em Python, índice de listas, arrays, etc. começam em 0.

```python
def indice_elementos(n, lista):
  for i in range(len(lista)):
    if lista[i] == n:
      return i
  return -1

```

```python
indice_elementos(-1, [4, 2, 9, 3, 5, 9, 6, 1])
```

(10) Implemente uma função que receba um número n e retorna a maior potência de 2 maior ou igual a n. Por exemplo: pot2(14) retornará 16, pot2(42), retornará 64.


```python
def pot2(n):
    potencia = 1
    while potencia < n:
        potencia *= 2
    return potencia
```

(11) Implemente uma função que dado um número natural maior do que 1, decomponha esse número em fatores primos. Por exemplo: se o argumento de entrada for 36, a saída deverá ser: [2,2,3,3], porque 2x2x3x3=36

```python
def decompor_em_fatores_primos(n):
    fatores = []
    divisor = 2
    while divisor * divisor <= n:
        if n % divisor:
            divisor += 1
        else:
            n //= divisor
            fatores.append(divisor)
    if n > 1:
        fatores.append(n)
    return fatores
```

```python
decompor_em_fatores_primos(36)
```

(12) Implemente a função maior N(lista, N) que recebe uma lista de numeros quaisquer, um valor N, 1 <= N <= len(lista) e retorne o Nenesimo maior valor na lista de números. Por exemplo, se N for 1, retorna o maior valor na lista, se N for 2, retorna o segundo maior valor na lista, etc.
Exemplo:
lista= [5,-1,7,0,-3,9]
N=2
  print(f'Em {lista} o {N}o. maior da lista é {maiorN(lista, N)}')

```python
def maiorN(lista, N):
    lista_ordenada = sorted(lista, reverse=True)
    return lista_ordenada[N-1]
```

```python
lista = [5,-1,7,0,-3,9]
N = 2
print(f'Em {lista} o {N}o. maior da lista é {maiorN(lista, N)}')
```

(13) Implemente uma função recursiva div(m, n), recebe como argumento dois números naturais m e n e devolve o resultado da divisão inteira de m por n. Não pode recorrer as operações aritméticas de multiplicação, divisão inteira e resto da divisão inteira.

```python
def div(m, n):
    if m < n:
        return 0

    return 1 + div(m - n, n)
```

```python
m = 10
n = 5
resultado = div(m, n)
print(f"A divisão inteira de {m} por {n} é {resultado}")
```

(15) Para calcular a raiz quadrada de um numero N > 0 podemos aplicar a seguinte formula iterativa:
Xn+1= 1/2(Xn+N/Xn)
onde Xn representa a raiz quadrada do número calculada na iteração n. As iterações terminam quando |N-Xn*Xn|<10**-9. Assuma X1=1 e implemente a função raiz (N, show=False). Note o parâmetro padrão (defaut) show. Implemente a função de forma que apenas quando show=True todos os valores estimados da raiz quadrada sejam impressos. Quando show=False, ou for ausente na chamada da função, nada será impresso. Em ambos os casos, o valor estimado da raiz quadrada será retornado pela função.


```python
def raiz_bissec(N, show=False):
    a, b = 0, N

    while True:
        Xn = (a + b) / 2
        if abs(N - Xn * Xn) < 1e-9:
            break

        if Xn * Xn < N:
            a = Xn
        else:
            b = Xn

        if show:
            print(f"Xn = {Xn:.9f}")

    return Xn
```

```python
numero = 25
resultado = raiz_bissec(numero, show=True)
print(f"Raiz quadrada de {numero} é aproximadamente {resultado:.9f}")
```
