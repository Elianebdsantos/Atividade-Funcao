```python
#FUNÇÃO RECURSIVA PARA CALCULO FATORIAL
# Fatorial > 0  ---- Fatoria = 0 é 1
def fatorial(n):
  if n == 1:
    return 1
  else:
    return n * fatorial(n-1)
resultado= fatorial(5)
print('O fatorial de 5 é: ', resultado)
```

```python
#FUNÇÃO ITERATIVA RECURSIVA PARA CALCULO DE FATORIAL

def fatorial(n):
    fatorial=1
    for i in range(2, n + 1):
        fatorial *=i
    return fatorial

resultado= fatorial(5)
print("O fatorial de 5 é:", resultado)
```

```python

#funçao iterativa potencia (a,b)
#potencia (a,b) = a*a*....*a, b vezes

def potenciaI(a,b):
    pot=1
    for i in range(b):
        pot = pot * a
    return pot

print(potenciaI(5,3))
```

```python
def estima_pi(n):
    """
    Calcula uma aproximação para o valor de π usando a série de Leibniz.

    Args:
        n (int): Número de parcelas (termos) usadas no cálculo.

    Returns:
        float: Aproximação para o valor de π.
    """
    pi_aproximado = 0
    sinal = 1  # Alterna o sinal dos termos

    for i in range(1, n + 1):
        termo = 1 / (2 * i - 1)  # Calcula o termo atual
        pi_aproximado += sinal * termo
        sinal *= -1  # Alterna o sinal para o próximo termo

    return 4 * pi_aproximado

# Exemplo de uso:
n_termos = 1000  # Escolha o número de termos desejado
aproximacao_pi = estima_pi(n_termos)
print(f"Aproximação para π com {n_termos} termos: {aproximacao_pi:.10f}")
```

```python
#Algoritmo para cálculo de raiz quadrada
def raiz(N):
  """
  Calcule a raiz quadrada de um número N
  Usando o algoritmo:
  x_(n+1) = 1/2 * (x_n + N/x_n)
  """
  x= 1
  for n in range(100):
    x= 1/2 * (x+ N/x)
  return x
```

```python
#solução da equação do 2º grau
def equacao_2_grau(a,b,c):
  """Função para retornar solução de
  equação do 2º grau

  Argumentos:
  a. coeficiente de x **2.
  b. coeficiente de x**1.
  c. coeficiente de x**0

  Retorna:
  Solução da equação:
  a * x**2 + b * x + c = 0
  (apenas quando b**2 - 4 *a*c > = 0)
  """
  delta= raiz(b**2 -4 * a * c)
  if delta < 0:
    print(f'Delta ({delta}) e menor do que zero!')
    return None
  r_delta = raiz(delta)
  x1 = (-b - r_delta)/(2 * a)
  x2 = (-b + r_delta)/(2 * a)
  return x1, x2
```

```python
# x-2)*(x-3) = 0
# x**2 - 5 * x + 6 = 0
a, b,c = 1, -5, 6
equacao_2_grau(a,b,c)
```

```python
#Função recursiva versus iterativa Implemente as funções soma_recursiva(n) e soma_iterativa(n) que recebem o número inteiro n, n > 0, e retornam a soma dos números inteiros de 1 até n, isto é, retornam 1 + 2 + 3 + ⋯ + n

#Nota: use boa prática e inclua documentação com docstrings.
#Dica: se n = 100 o resultado deve ser 5050.

#s(n) = n + s(n-1)

def soma_r(n):
  if n == 1:
    return 1
  return n + soma_r(n-1)
```

```python
#Fibonatti
#Crie uma função recursiva que retorna o n-ésimo termo da sequência de Fibonacci. A sequência de Fibonacci é da forma: 1, 1, 2, 3, 5, 8, 13, 21, 34, ...
#Onde o termo corrente é a soma dos dois termos anteriores, isto é, fibonacci(n) = fibonacci(n-1) + fibonacci(n-2).

def fibo(n):
  if n <= 1:
    return n
  else:
    return fibo(n-1) + fibo(n-2)
```

```python
#outra forma de fazer mais simplificada
def fibo(n):
  if n <= 1: return n
  return fibo(n-1) + fibo(n-2)
```
